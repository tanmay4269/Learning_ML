# Associative memory

```python
import numpy as np
import matplotlib.pyplot as plt

n = 7
v_s = np.full((n, n), -1)  # this is the memory we wanna achieve
# defining v_s
for i in range(n):
    for j in range(n):
        if i == j or i + j == n - 1:
            v_s[i, j] = 1

print(v_s)

image = v_s.reshape((n,n))
plt.imshow(image)
plt.axis('off')
plt.show()
```

```python
v = np.random.choice([-1, 1], (n, n))

plt.imshow(v)
plt.axis('off')
plt.show()
```
