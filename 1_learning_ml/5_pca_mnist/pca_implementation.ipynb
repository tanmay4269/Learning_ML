```python
import idx2numpy
import matplotlib.pyplot as plt 
import numpy as np
from numpy import linalg as LA
```

```python
train_images_file = 'train-images.idx3-ubyte'
train_images_array = idx2numpy.convert_from_file(train_images_file)

train_labels_file = 'train-labels.idx1-ubyte'
train_labels_array = idx2numpy.convert_from_file(train_labels_file)
```

```python
print(train_images_array.shape)
print(train_labels_array.shape)
```

```python
plt.imshow(train_images_array[0, :, :], cmap="gray")
plt.axis("off")
plt.show()
```

```python
# Extracting all images of the digit 5.
fav_digit = []

for i, digit in enumerate(train_labels_array):
    if digit == 5:
        fav_digit.append(train_images_array[i, :, :])

fav_digit = np.array(fav_digit)

fav_digit.shape
```

```python
avg_digit = np.average(fav_digit, axis=0)
plt.imshow(avg_digit, cmap='gray')
plt.axis('off')
plt.show()
```

```python
plt.imshow(fav_digit[1, :, :], cmap="gray")
plt.axis("off")
plt.show()
```

```python
from sklearn.decomposition import PCA

pca_784 = PCA(n_components=784)
pca_784.fit(fav_digit.reshape(-1, 28*28))

plt.grid()
plt.plot(np.cumsum(pca_784.explained_variance_ratio_ * 100))
plt.xlabel('Number of components')
plt.ylabel('Explained variance')
plt.show()
```

`explained_variance_ratio_` property of PCA is (ith eigen value) / (sum of all eigen values)

```python
val = np.cumsum(pca_784.explained_variance_ratio_ * 100)

for i in range(784):
    if val[i] > 99:
        print(i)
        break

print(pca_784.explained_variance_ratio_.shape)
print(val.shape)
```

```python
mnist = fav_digit.reshape(-1, 28*28)

pca_10 = PCA(n_components=10)
mnist_pca_10_reduced = pca_10.fit_transform(mnist)
mnist_pca_10_recovered = pca_10.inverse_transform(mnist_pca_10_reduced)

image_pca_10 = mnist_pca_10_recovered[1,:].reshape([28,28])
plt.imshow(image_pca_10, cmap='gray_r')
plt.show()
```
